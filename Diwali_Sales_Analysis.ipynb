```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding= 'unicode_escape')
```

```python
df.head(15)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df['Amount']= df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age', 'Orders']].describe()
```

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
bx = sns.barplot(x= 'Gender', y= 'Amount',data= sales_gen)

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
df.columns
```

```python
cx = sns.countplot(data = df, x= 'Age Group', hue= 'Gender')

for bars in cx.containers:
    cx.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending=False)
dx = sns.barplot(x= 'Age Group', y= 'Amount',data= sales_age)

for bars in dx.containers:
    dx.bar_label(bars)
```

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(7)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x= 'State', y = 'Orders')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(7)

sns.set(rc={'figure.figsize':(15,5)})
ex = sns.barplot(data = sales_state, x= 'State', y = 'Amount')

for bars in ex.containers:
    ex.bar_label(bars)    
```

```python
df.columns
```

```python
fx = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in fx.containers:
    fx.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
ex = sns.barplot(data = sales_state, x= 'Marital_Status', y = 'Amount', hue='Gender')

for bars in ex.containers:
    ex.bar_label(bars)    
```
